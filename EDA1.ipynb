## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
#creating a duplicate of the original dataset before making any changes in it
hotel_df1 = hotel_df.copy()

hotel_df1.columns
```

```python
#Replacing null values in children column with 0 assuming that family had 0 children
#Droping company and agent columns

hotel_df1['children'].fillna(0,inplace=True)
hotel_df1.drop(columns=['agent','company'],inplace=True)
```

```python
#Replacing null values in country column as 'others'.
hotel_df1['country'].fillna('others',inplace=True)


#Checking for null values after replacing them
hotel_df1.isnull().sum()
```

```python
#droping rows where no outlets,children and babies are available because no booking has made that day.
no_guest=hotel_df1[hotel_df1['adults']+hotel_df1['babies']+hotel_df1['children']==0]
hotel_df1.drop(no_guest.index,inplace=True)
```

```python
# We will do some feature engineering
#adding some new columns to make our data analysis ready
hotel_df1['total_people']=hotel_df1['adults']+hotel_df1['babies']+hotel_df1['children']
hotel_df1['total_stay']=hotel_df1['stays_in_weekend_nights']+hotel_df1['stays_in_week_nights']
```

```python
#having a final look to check if our dataset is ready to analyse
hotel_df1.head(2)
# dropping adults,babies and children
hotel_df1.drop(columns=['adults','babies','children'],inplace=True)

```

```python
#checking the final shape of the dataset
print(f'Final shape of the dataset is' ,(hotel_df1.shape))
```

```python
#changing datatype
hotel_df1['hotel']=hotel_df1['hotel'].astype('category')
hotel_df1['is_canceled']=hotel_df1['is_canceled'].astype('category')
hotel_df1['arrival_date_year']=hotel_df1['arrival_date_year'].astype('category')
hotel_df1['meal']=hotel_df1['meal'].astype('category')
hotel_df1['distribution_channel']=hotel_df1['distribution_channel'].astype('category')
hotel_df1['is_repeated_guest']=hotel_df1['is_repeated_guest'].astype('category')
hotel_df1['reserved_room_type']=hotel_df1['reserved_room_type'].astype('category')
hotel_df1['assigned_room_type']=hotel_df1['assigned_room_type'].astype('category')
hotel_df1['deposit_type']=hotel_df1['deposit_type'].astype('category')
hotel_df1['customer_type']=hotel_df1['customer_type'].astype('category')
hotel_df1['required_car_parking_spaces']=hotel_df1['required_car_parking_spaces'].astype('category')
hotel_df1['total_of_special_requests']=hotel_df1['total_of_special_requests'].astype('category')
hotel_df1['reservation_status']=hotel_df1['reservation_status'].astype('category')
hotel_df1['total_people']=hotel_df1['total_people'].astype('int')

hotel_df1.info()
```

##Char-14-What is the adr across different months?

```python
# Grouping arrival_month and hotel on mean of adr
bookings_months=hotel_df1.groupby(['arrival_date_month','hotel'])['adr'].mean().reset_index()

# Creating a month list to order the months in ascending
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Creating a dataset of months, hotel and their adr
bookings_months['arrival_date_month']=pd.Categorical(bookings_months['arrival_date_month'],categories=months,ordered=True)

# Sorting the months
bookings_months=bookings_months.sort_values('arrival_date_month')
bookings_months
```

```python
# Setting the chart size
plt.figure(figsize=(15,5))

# Plotting the values in a line chart
sns.lineplot(x=bookings_months['arrival_date_month'],y=bookings_months['adr'],hue=bookings_months['hotel'])

# Setting the labels and title
plt.title('ADR across each month', fontsize=20)
plt.xlabel('Month Name', fontsize=12)
plt.ylabel('ADR', fontsize=12)

# Show chart
plt.show()
```

1. Why you used the specific chart?

Line graphs are used to track changes over short and long periods of time.

2. What are the insight(s) from the above chart?

**City Hotel :** It is clear that City Hotel generates more revenue in May months in comparison to other months.

**Resort Hotel :** Resort Hotel generates more revenue in between July and August months.
